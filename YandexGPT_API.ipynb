```python
text_to_save = answer["result"]["alternatives"][0]["message"]["text"]
```

```python
from openpyxl import Workbook
```

```python
wb = Workbook()
ws = wb.active
ws.title = "Текст ответа"
```

```python
for i, line in enumerate(text_to_save.split('\n'), start=1):
     ws.cell(row=i, column=1, value=line)
ws.column_dimensions['A'].width = 100
```

```python
wb.save("yandexGPT_text.xlsx")
print(f"Текст успешно сохранен в файл 'yandexGPT_text.xlsx'")
```
